```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from joblib import dump, load
```

```python
X, y = make_regression(n_samples=2000, n_features=3, noise=0.1)
```

```python
model = LinearRegression().fit(X, y)
dump(model, "linear_model.joblib")
model_loaded = load("linear_model.joblib")
```

```python
print(model_loaded.predict(X[:5]))
```

Man skapar data med 2000 datapunkter, varje datapunkt har 3 features. En linjär modell skapas och tränar på datan som vi tidigare skapat och lär sig ett samband. Med dump model sparas modellen för att kunna använda den senare utan att träna den igen. Sedan läser man modellen och gör prediktioner på de första 5 datapunkterna som blir outputen.
