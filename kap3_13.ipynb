```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('hr_employee_data.xlsx')
```

```python
plt.figure()
plt.boxplot(df['satisfaction_level'])
plt.title("Box Plot of Satisfaction Level")
```

```python
df.describe()
```

```python
plt.figure()
plt.boxplot(df['last_evaluation'])
plt.title("Box Plot of Last Evaluation")
```

```python
counts = df["Department"].value_counts()
```

```python
counts.plot(kind="bar")
plt.xlabel("Department")
plt.ylabel("Antal")
plt.title("Antal per department")
plt.show()
```

```python
counts.salary = df["salary"].value_counts()
```

```python
counts.salary.plot(kind="bar")
plt.xlabel("Lön")
plt.ylabel("Antal")
plt.title("Fördelningen av löner")
plt.show()
```

```python
df["average_montly_hours"].plot(kind="hist")
plt.xlabel("Antal timmar per månad")
plt.ylabel("Antal")
plt.title("Fördelning av arbetstimmar")
plt.show()
```

```python
df["average_montly_hours"].plot(kind="box")
plt.title("Boxplot av arbetstimmar")
plt.show()
```

```python
df.groupby("Department")["average_montly_hours"].mean().plot(kind="bar")
plt.ylabel("Medel timmar")
plt.title("Arbetstimmar per department")
plt.show()
```
